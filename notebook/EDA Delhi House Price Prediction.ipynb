### 2.6 Explore the different categories in different columns

```python
# One way to find unique values for a specific column is to call .unique() directly on that column.
# For example, to find unique values in the total_floors column:
df["total_floors"].unique()
```

```python
# Alternatively, you can create a function to get the unique values for each column.
# Note: You can implement this in any way you prefer. I’ve defined the function this way because it suits my style.
# However, feel free to use a different approach if it makes you more comfortable.
# Just remember that others reading or reviewing your work should also be able to understand your code.

def get_unique(data, columns: list):
    """
    Prints the unique values in each specified column of the dataset.
    
    Parameters:
    - data: The DataFrame containing the dataset.
    - columns: A list of column names for which to display unique values.
    
    Output:
    - Prints each column name followed by the list of unique values in that column.
    """
    for column in columns:
        print(f"Unique values in '{column}' column:\n{data[column].unique()}")
        print("=" * 40)

columns = ["city", "property_type", "bedroom_num", "bathroom_num", "balcony_num", "furnished", "age", "total_floors"]
get_unique(df, columns=columns)
```

### 3.1 Data Distribution Analysis

```python
# histogram for numerical caategory distribution
numeric_cols = ['price', 'area', 'price_per_sqft', 'bedroom_num', 'bathroom_num', 'balcony_num', 'age']
df[numeric_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Distributions of Numerical Features")
plt.show()

```

```python
# Histogram for price distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# KDE for area
plt.figure(figsize=(8, 6))
sns.kdeplot(df['area'], fill=True)
plt.title("Area Distribution")
plt.xlabel("Area")
plt.ylabel("Density")
plt.show()
```

After visualizing the distributions of numerical features in the dataset, we observed the following patterns:

1. **Price**: The distribution of property prices is heavily **right-tailed** (positively skewed). This skewness indicates that while most properties are priced in a lower range, there are a few properties with exceptionally high prices, pulling the mean to the right.

2. **Area**: The area of properties also shows a **right-skewed distribution**. This pattern suggests that most properties fall within a typical size range, but a few properties with much larger areas extend the distribution towards higher values.

3. **Price per Square Foot**: This feature is similarly **right-skewed**, meaning that while the majority of properties have a standard price per square foot, a small subset of properties are much more expensive on a per-unit basis.

4. **Age**: The property age distribution is **right-tailed** as well. Most properties in this dataset are relatively new, with fewer older properties represented, which could reflect recent construction trends.

5. **Bedroom Number, Bathroom Number, and Balcony Number**: These features show slight **right-skewness**, although less pronounced compared to the other variables. Most properties have a typical number of bedrooms, bathrooms, and balconies, with fewer properties having more rooms or balconies.

**Interpretation**: The skewness in the distributions for features like `price`, `area`, and `price_per_sqft` indicates the presence of high-value outliers, which could impact model performance if not handled appropriately. For `age`, the distribution implies that newer properties dominate the dataset. Right-skewness in `bedroom_num`, `bathroom_num`, and `balcony_num` is expected, as most properties are likely designed with standard configurations.


### 3.2 Correlation Analysis

```python
# lets see the correlation of out numeric column
# you can directly call pandas .corr() function setting numetic_only param True
# you can use seaborn heatmap to visualize the correlation

plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix of Numerical Columns")
plt.show()
```

***Correlation Insights***

Here, we observe that our target column (`price`) has a **high correlation** with `area` (correlation = 0.71), and **moderate correlations** with `price_per_sqft`, `bedroom_num`, and `bathroom_num`, with correlation values of 0.66, 0.54, and 0.53, respectively. 

In contrast, `balcony_num` and `age` have **weak positive correlations** with `price`, with values of 0.068 and 0.034.

### 3.3 Exploring `property_type`, `furnished`,  and `city`

```python
def plot_bar_price_by_hue(df, x, y, hue, title, rotation=45):
    """
    This function create a barplot.

    parma:
    df = dataframe
    x = x value
    y = y value
    hue
    title
    rotation(default 45)
    """
    plt.figure(figsize=(12, 6))
    sns.barplot(x=x, y=y, hue=hue, data=df)
    plt.title(title)
    plt.xticks(rotation=rotation)
    plt.legend(title=hue.capitalize())
    plt.show()


# Price by Property Type with Furnished Hue
plot_bar_price_by_hue(df, 'property_type', 'price', 'furnished', "Price by Property Type (Furnished)")

# Price by Property Type with City Hue
plot_bar_price_by_hue(df, 'city', 'price', 'property_type', "Price by City (Property_type)")
```

***Barplot Insights***

In the first plot, **Price by Property Type (Furnished)**, we observe that **Independent Floor** with **Unfurnished** status has a relatively high price compared to other property types. Upon further review, we can see that **Villa** has the second highest price, while **Studio Apartment** has the lowest price and is only available as **Furnished**. Overall, the plot clearly indicates that, regardless of property type, furnished properties have higher prices than semi-furnished and unfurnished ones.

In the second plot, **Price by City (Property Type)**, we observe that **South Mumbai** has the highest price compared to other areas in Mumbai. Additionally, **Studio Apartments** are only available in **Central Mumbai** and **Western Mumbai**. We can also observe that **Independent Floors** are unique to Mumbai, making it the only city among South, Central, and Western Mumbai that has different kinds of property types available: Apartment, Villa, Independent House, and Independent Floor.


### 3.4 Statistic Test For Categorical Feature

```python
from scipy.stats import f_oneway

# Example: ANOVA test for price differences by property type
property_types = df['property_type'].unique()
price_by_type = [df[df['property_type'] == p]['price'] for p in property_types]
f_stat, p_val = f_oneway(*price_by_type)
print(f"ANOVA Test Result for Property Type: F={f_stat}, p={p_val}")

# Repeat similarly for `city` if there are fewer unique values, as high cardinality can affect ANOVA.
```
